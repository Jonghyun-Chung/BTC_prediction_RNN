```python
import re

new_data = data.copy()
rep = {"Jan ": "01-", "Feb ": "02-","Mar ": "03-","Apr ": "04-","May ": "05-","Jun ": "06-","Jul ": "07-","Aug ": "08-","Sep ": "09-","Oct ": "10-","Nov ": "11-","Dec ": "12-",} 

rep = dict((re.escape(k), v) for k, v in rep.items()) 
pattern = re.compile("|".join(rep.keys()))
pattern2 = re.compile(', ')

for nd in new_data:
    nd[0] = pattern.sub(lambda m: rep[re.escape(m.group(0))], nd[0])
    nd[0] = pattern2.split(nd[0])
    nd[0] = nd[0][1] + '-'+ nd[0][0]
    
    nd[1] = nd[1].replace('$', '')
    nd[2] = nd[2].replace('$', '')
    nd[3] = nd[3].replace('$', '')
    nd[4] = nd[4].replace('$', '')
    nd[5] = nd[5].replace('$', '')
    nd[6] = nd[6].replace('$', '')

new_data
```

```python
import csv
fields = ['date', 'open', 'high', 'low', 'close', 'volume', 'market_cap'] 
rows = new_data

with open('Bitcoin_transactions.csv', 'w') as f: 
    write = csv.writer(f) 
    write.writerow(fields) 
    write.writerows(rows) 
```
